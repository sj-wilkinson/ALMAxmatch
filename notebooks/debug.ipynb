```python
from astroquery.alma import Alma
from astropy import coordinates
from astropy import units as u
```

```python
# one galaxy from ALMaQueSt
gal = coordinates.SkyCoord(49.82143689*u.deg, -0.96964645*u.deg, frame='icrs')
```

```python
alma = Alma()
payload = dict() ## payload is a dictionary of search terms
#payload['ra_dec'] = gal.fk5.to_string(style = 'hmsdms', sep = ':') # old method that doesn't work!
payload['s_ra'] = gal.ra.to_string()
payload['s_dec'] = gal.dec.to_string()
data = alma.query_region(gal, maxrec = 10, public = True, science = True, radius = 30*u.arcsec)
print(data) 
```
